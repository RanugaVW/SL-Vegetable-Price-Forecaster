# USD Exchange Rate Impact Analysis

This notebook investigates the effect of the USD Exchange Rate on the Farmer Price and the Retail Price over the duration of our dataset. We aim to show this relationship using correlation calculations and dual-axis line charts.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Set up paths and load data
data_path = r"C:\Users\Ranuga\Data Science Project\3. Data Preprocessing\3.7 - Combining Datasets\Outputs\Final_Combined_data.csv"
df = pd.read_csv(data_path)

# Ensure chronologically sorted format
df['week_num'] = df['week'].str.extract(r'(\d+)').astype(int)
df = df.sort_values(['year', 'week_num'])

# Create a display string for time plotting
df['Year_Week_Str'] = df['year'].astype(str) + "-W" + df['week_num'].astype(str).str.zfill(2)

df.head()
```

## Aggregate Data

To get a clearer view of the macro trends, we will group the data averaging it across all vegetable types for each given week.

```python
# Group across all vegetables per week for a macro view
overall_trend = df.groupby(['year', 'week_num', 'Year_Week_Str'])[['retail_price', 'mean_farmer_price', 'usd_exchange_rate']].mean().reset_index()

overall_trend.head()
```

## Dual-Axis Visualization

The goal is to map the retail/farmer prices on the primary Y-axis, while charting the USD Exchange against a secondary Y-axis to understand trend alignments.

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot Vegetable Prices (Primary Axis)
ax1.plot(overall_trend['Year_Week_Str'], overall_trend['retail_price'], color='green', label='Avg Retail Price (LKR)')
ax1.plot(overall_trend['Year_Week_Str'], overall_trend['mean_farmer_price'], color='orange', label='Avg Farmer Price (LKR)')
ax1.set_xlabel('Time (Year-Week)', fontweight='bold')
ax1.set_ylabel('Vegetable Price (LKR)', color='black', fontweight='bold')

# Dynamically arrange x-ticks to prevent overlapping
ticks = np.arange(0, len(overall_trend), max(1, len(overall_trend)//15))
ax1.set_xticks(ticks)
ax1.set_xticklabels(overall_trend['Year_Week_Str'].iloc[ticks], rotation=45)

# Plot USD Exchange Rate (Secondary Axis)
ax2 = ax1.twinx()
ax2.plot(overall_trend['Year_Week_Str'], overall_trend['usd_exchange_rate'], color='blue', linestyle='--', label='USD Exchange Rate')
ax2.set_ylabel('USD Exchange Rate (LKR)', color='blue', fontweight='bold')

plt.title('Overall Market Prices vs USD Exchange Rate', fontsize=14, fontweight='bold')
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

## Correlation Assessment

```python
# Calculate Pearson correlations
corr_retail = overall_trend['usd_exchange_rate'].corr(overall_trend['retail_price'])
corr_farmer = overall_trend['usd_exchange_rate'].corr(overall_trend['mean_farmer_price'])

print(f"Pearson Correlation (USD vs Avg Retail Price): {corr_retail:.4f}")
print(f"Pearson Correlation (USD vs Avg Farmer Price): {corr_farmer:.4f}")
```
