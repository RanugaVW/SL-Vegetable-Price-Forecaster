```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../main data set/main_data_cleaned.csv')
```

```python
data.head()
```

```python
data = data[['Year_Week', 'lanka_auto_diesel_price', 'usd_exchange_rate']]
```

```python
data = data.drop_duplicates().reset_index(drop=True)
data.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
plt.plot(data['Year_Week'], data['lanka_auto_diesel_price'], label='Lanka Auto Diesel Price', color='blue')
plt.plot(data['Year_Week'], data['usd_exchange_rate'], label='USD Exchange Rate', color='orange')
plt.xlabel('Year_Week')
plt.ylabel('Value')
plt.title('Year_Week vs Lanka Auto Diesel Price and USD Exchange Rate')
plt.legend()
plt.xticks(ticks=range(0, len(data), 26), labels=data['Year_Week'][::26], rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
data[['lanka_auto_diesel_price', 'usd_exchange_rate']] = scaler.fit_transform(
    data[['lanka_auto_diesel_price', 'usd_exchange_rate']]
)
data.head()
```

```python
plt.figure(figsize=(14, 6))
plt.plot(data['Year_Week'], data['lanka_auto_diesel_price'], label='Lanka Auto Diesel Price (Normalized)', color='blue')
plt.plot(data['Year_Week'], data['usd_exchange_rate'], label='USD Exchange Rate (Normalized)', color='orange')
plt.xlabel('Year_Week')
plt.ylabel('Normalized Value')
plt.title('Year_Week vs Normalized Lanka Auto Diesel Price and USD Exchange Rate')
plt.legend()
plt.xticks(ticks=range(0, len(data), 26), labels=data['Year_Week'][::26], rotation=45)
plt.tight_layout()
plt.show()
```

```python
correlation = data['lanka_auto_diesel_price'].corr(data['usd_exchange_rate'])
print("Correlation between Lanka Auto Diesel Price and USD Exchange Rate:", correlation)
```
